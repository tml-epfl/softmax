```python
import numpy as np

softmax = np.load("data/attention_stats_detailed_softmax.npy", allow_pickle=True)
sigmoid = np.load("data/attention_stats_detailed_sigmoid.npy", allow_pickle=True)
```

```python
# The data is loaded as a 0-dimensional numpy array containing a dictionary
# To access the nested data, first extract the dictionary using .item()
softmax_data = softmax.item()
sigmoid_data = sigmoid.item()

# Now you can access the nested dictionaries
softmax_max = softmax_data["softmax"]["max_proportion"]
sigmoid_max = sigmoid_data["sigmoid"]["max_proportion"]

softmax_first = softmax_data["softmax"]["first_token_proportion"]
sigmoid_first = sigmoid_data["sigmoid"]["first_token_proportion"]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ICML-style settings
plt.rcParams.update({
    'font.family': 'serif',
    'font.serif': ['Times', 'Times New Roman', 'DejaVu Serif'],
    'font.size': 9,
    'axes.labelsize': 10,
    'axes.titlesize': 10,
    'xtick.labelsize': 8,
    'ytick.labelsize': 8,
    'legend.fontsize': 8,
    'figure.dpi': 150,
    'savefig.dpi': 300,
    'savefig.bbox': 'tight',
    'savefig.pad_inches': 0.02,
    'axes.spines.top': False,
    'axes.spines.right': False,
})

fig, ax = plt.subplots(figsize=(3.25, 1.8))

sns.histplot(softmax_max.flatten(), label='Softmax', ax=ax, alpha=0.8, stat='count', bins=20, color=sns.color_palette("Set2", 4)[2])
sns.histplot(sigmoid_max.flatten(), label='Sigmoid (w/o norm.)', ax=ax, alpha=0.8, stat='count', bins=20, color=sns.color_palette("Set2", 4)[1])

ax.set_xlabel('avg. sparsity score')
ax.set_ylabel('# heads')
ax.legend(frameon=False)

plt.tight_layout()

plt.savefig('visualizations/fig/sink_scores.pdf')
```
